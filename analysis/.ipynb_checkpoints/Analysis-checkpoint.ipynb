```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = pd.read_csv('../data/phd.csv', sep=';')
```

```python
data.head()
```

```python
data = data[data['Is PhD Student'] == 1.0]
data = data.drop(['Is PhD Student', 'Is PhD Student From South'], axis=1)
data.shape
```

```python
data.iloc[:,9:-1] = data.iloc[:,9:-1].replace({np.nan: False, 1.0: True})
data['HDR'] = data['HDR'].replace({0.0: False, 1.0: True})
data['Sex'] = data['Sex'].replace({'F': 'Female', 'H': 'Male'})
data['Year of Birth'] = data['Year of Birth'].astype('Int64')
data.head()
```

```python
data.to_csv('../dphd.csv', index=False)
```
